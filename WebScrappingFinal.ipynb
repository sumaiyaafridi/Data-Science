```python
from bs4 import BeautifulSoup
import requests
```

```python
print("Skills that I am not familiar with")
unskill=input('>')
print(f'Filtering out:{unskill}')
def find_datajobs():
    html=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=Data+Analysis&txtLocation=').text
    soup=BeautifulSoup(html,'lxml')
    jobs=soup.find_all('li', class_='clearfix job-bx wht-shd-bx')
    for index, job in enumerate(jobs):
        date=job.find('span',class_='sim-posted').span.text
        if 'few' in date:
            name=job.find('h3', class_="joblist-comp-name").text.replace(' ','')
            skills=job.find('span',class_='srp-skills').text.replace(' ','')
            link=job.header.h2.a['href']
            if unskill not in skills:
                with open(f'posts/{index}.txt','w') as f:
                    f.write(f"Company Name: {name.strip()}\n")
                    f.write(f"Required Skills:{skills.strip()}\n")
                    f.write(f"Date:{date}\n")
                    f.write(f"Apply: {link}\n")

                    print('')
            print(f'File saved: {index}')
```

```python
import time
if __name__=='__main__':
    while True:
        find_datajobs()
        time_wait=10
        print(f'waiting {time_wait} seconds.........')
        time.sleep(time_wait * 1)
        
```
